```python
import pandas as pd
import numpy as np
```

```python
data = {
    "EmpID": ["E101", "E102", "E103", "E104", "E105"],
    "Name": ["Atul", "Raj", "Darpan", "Anmol", "Piyush"],
    "Age": [25, 30, 23, 22, 32],
    "Designation": ["Manager", "Clerk", "Analyst", "Clerk", "Manager"],
    "Salary": [56000, 25000, 35000, 28000, 58000],
    "Bonus": [15000, 7000, 9000, 13000, 12000]
}
```

```python
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
new_row = pd.DataFrame([{
    "EmpID": "E106",
    "Name": "GEET",
    "Age": 25,
    "Designation": "Analyst",
    "Salary": 70000,
    "Bonus": 15000
}])
```

```python
df = pd.concat([df, new_row], ignore_index=True)
print(df)
```

```python
# b) Add column "Shift" with "Morning"
df["Shift"] = "Morning"
```

```python
# c) Print details of youngest employee
youngest = df[df["Age"] == df["Age"].min()]
print("\nYoungest Employee:\n", youngest)
```

```python
# d) Print details of all clerks
clerks = df[df["Designation"] == "Clerk"]
print("\nAll Clerks:\n", clerks)
```

```python
# e) Print average Bonus
avg_bonus = df["Bonus"].mean()
print("Average Bonus:", avg_bonus)
```

```python
# f) Print total salary distributed
total_salary = df["Salary"].sum()
print("Total Salary Distributed:", total_salary)
```

```python
# g) Add column "City"
cities = ["Janak Puri", "Tilak Nagar", "Shiv Nagar", "Vikas Puri", "Janak Puri", "Tilak Nagar"]
df["City"] = cities
```

```python
# h) Delete record with EmpID 'E103' and 'E105'
df = df[~df["EmpID"].isin(["E103", "E105"])]
```

```python
# i) Rename column "Bonus" as "Incentive"
df = df.rename(columns={"Bonus": "Incentive"})
```

```python
print("\nFinal DataFrame:\n", df)
```

#Stationery DataFrame

```python
data = {
    "Count": [15, 20, 25, 35],
    "Price": [250, 200, 600, 900]
}
index = ["Pencil", "Pen", "Eraser", "Scale"]
```

```python
df = pd.DataFrame(data, index=index)
print("Original DataFrame:\n", df)
```

```python
# a) Change the Eraser count to 15 instead of 25
df.loc["Eraser", "Count"] = 15
print(df)
```

```python
# b) Rename the index "Scale" as "Wooden Scale"
df = df.rename(index={"Scale": "Wooden Scale"})
print(df)
```

```python
# c) Change all row labels to Product 1, Product 2, Product 3, Product 4
df.index = ["Product 1", "Product 2", "Product 3", "Product 4"]
print(df)
```

```python
# d) Change all column names as "Discount", "Total"
df.columns = ["Discount", "Total"]
print(df)
```

```python
# e) Add one row "Product 5" with value 10 and 100
df.loc["Product 5"] = [10, 100]
print(df)
```

```python
Products = df.drop(columns=["Total"])
print(Products)
```

#State-wise Production Data

```python
data2 = {
    "Fruits": [44, 11950, 7152, 140169, 7830, 113],
    "Pulses": [23, 818, 33, 3184, 931, 1],
    "Rice": [814, 1930, 11586, 13574, 7452, 2604],
    "Wheat": [5, 2737, 16440, 30056, np.nan, np.nan]
}

states = ["Tripura", "Gujrat", "Punjab", "UP", "AP", "Kerala"]
```

```python
df2 = pd.DataFrame(data2, index=states)
print("\nProduction DataFrame:\n", df2)
```

```python
# a) Total production for each state
total_per_state = df2.sum(axis=1)
print("\nTotal production for each state:\n", total_per_state)
```

```python
# b) Total production for each item type
total_per_item = df2.sum(axis=0)
print("\nTotal production for each item type:\n", total_per_item)
```

```python
# c) How many states produce Wheat? (non-NaN values)
wheat_states = df2["Wheat"].count()
print("\nNumber of states producing Wheat:", wheat_states)
```
